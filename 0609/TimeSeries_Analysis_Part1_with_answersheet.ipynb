## 3. ì‹œê³„ì—´ ë°ì´í„° ì¸ë±ì‹±ê³¼ ë¦¬ìƒ˜í”Œë§

### ğŸ’¡ ê°œë… (Concept)

> ì‹œê³„ì—´ ë°ì´í„°ì˜ í•µì‹¬ì€ 'ì‹œê°„'ì„ ê¸°ì¤€ìœ¼ë¡œ ë°ì´í„°ë¥¼ ìë¥´ê³ , ë³€í˜•í•˜ê³ , ì§‘ê³„í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤. ë‚ ì§œë¥¼ ë°ì´í„°í”„ë ˆì„ì˜ ì¸ë±ìŠ¤(index)ë¡œ ì„¤ì •í•˜ë©´ ê°•ë ¥í•œ ì‹œê³„ì—´ ë¶„ì„ ê¸°ëŠ¥ì„ í™œìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. 
> 
> **ë¦¬ìƒ˜í”Œë§(Resampling)** ì€ ì‹œê³„ì—´ ë°ì´í„°ì˜ ì‹œê°„ ë¹ˆë„ë¥¼ ë³€ê²½í•˜ëŠ” ê³¼ì •ì…ë‹ˆë‹¤. 
> 
> ì˜ˆë¥¼ ë“¤ì–´, ì¼ë³„(daily) ë°ì´í„°ë¥¼ ì£¼ë³„(weekly)ì´ë‚˜ ì›”ë³„(monthly) ë°ì´í„°ë¡œ ë³€í™˜í•˜ì—¬ í° íë¦„ì„ ë³´ê±°ë‚˜(ë‹¤ìš´ìƒ˜í”Œë§), ë¶„ ë‹¨ìœ„ ë°ì´í„°ë¥¼ ì´ˆ ë‹¨ìœ„ë¡œ ì„¸ë¶„í™”(ì—…ìƒ˜í”Œë§)í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### `set_index`ì™€ `resample`: ì‹œê°„ ë‹¨ìœ„ë¡œ ë°ì´í„° ì§‘ê³„í•˜ê¸°
> `set_index`ë¡œ ë‚ ì§œ ì»¬ëŸ¼ì„ ì¸ë±ìŠ¤ë¡œ ì§€ì •í•œ í›„, `resample()` ë©”ì†Œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì›í•˜ëŠ” ì‹œê°„ ë‹¨ìœ„ë¡œ ë°ì´í„°ë¥¼ ê·¸ë£¹í™”í•˜ê³ , `.mean()`, `.sum()`, `.last()` ë“±ê³¼ ê°™ì€ ì§‘ê³„ í•¨ìˆ˜ë¥¼ ì ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

```python
# ì‹¤ìŠµì„ ìœ„í•œ ì˜ˆì œ ë°ì´í„° ìƒì„±
date_rng = pd.date_range(start='2023-01-01', periods=100, freq='D')
np.random.seed(42)
data = np.random.randint(50, 200, size=(100,))
ts_df = pd.DataFrame(data, index=date_rng, columns=['sales'])

print("--- ì›ë³¸ ë°ì´í„° (ì¼ë³„ íŒë§¤ëŸ‰) ---")
print(ts_df.head())

# ì£¼ë³„(Weekly) í‰ê·  íŒë§¤ëŸ‰ ê³„ì‚° (ë‹¤ìš´ìƒ˜í”Œë§)
# 'W'ëŠ” ì£¼ì˜ ë§ˆì§€ë§‰ ë‚ (ì¼ìš”ì¼)ì„ ê¸°ì¤€ìœ¼ë¡œ ì§‘ê³„í•©ë‹ˆë‹¤.
weekly_mean_sales = ts_df['sales'].resample('W').mean()

print("--- ì£¼ë³„ í‰ê·  íŒë§¤ëŸ‰ ---")
print(weekly_mean_sales.head())
```

```python
# ì›”ë³„(Monthly) ì´ íŒë§¤ëŸ‰ ê³„ì‚° (ë‹¤ìš´ìƒ˜í”Œë§)
# 'MS'ëŠ” ì›”ì˜ ì‹œì‘ì¼(1ì¼)ì„ ê¸°ì¤€ìœ¼ë¡œ ì§‘ê³„í•©ë‹ˆë‹¤.
monthly_sum_sales = ts_df['sales'].resample('MS').sum()

print("--- ì›”ë³„ ì´ íŒë§¤ëŸ‰ ---")
print(monthly_sum_sales.head())

# Plotlyë¥¼ ì‚¬ìš©í•œ ì‹œê°í™”
fig = px.bar(monthly_sum_sales, 
             title='ì›”ë³„ ì´ íŒë§¤ëŸ‰', 
             labels={'value':'ì´ íŒë§¤ëŸ‰', 'index':'ì›”'})
fig.show()
```

### âœï¸ ì—°ìŠµ ë¬¸ì œ (Practice Problems)

1. ìœ„ì—ì„œ ìƒì„±í•œ `ts_df` ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬, **ë¶„ê¸°ë³„(Quarterly) íŒë§¤ëŸ‰ì˜ ìµœëŒ“ê°’**ì„ ê³„ì‚°í•˜ê³  ì¶œë ¥í•´ë³´ì„¸ìš”. (`freq` ì½”ë“œëŠ” 'Q'ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤.)

## 4. ì´ë™ì°½ í•¨ìˆ˜ (Window Functions)

### ğŸ’¡ ê°œë… (Concept)

> ì‹œê³„ì—´ ë°ì´í„°ì˜ ë…¸ì´ì¦ˆ(noise)ë¥¼ ì¤„ì´ê³  ì¥ê¸°ì ì¸ ì¶”ì„¸ë¥¼ íŒŒì•…í•˜ê¸° ìœ„í•´ **ì´ë™ì°½(Window)** í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤. ì´ëŠ” ì¼ì • í¬ê¸°ì˜ ì°½(window)ì„ ë°ì´í„° ìœ„ì—ì„œ í•œ ì¹¸ì”© ì´ë™ì‹œí‚¤ë©´ì„œ ì°½ ë‚´ë¶€ì˜ ë°ì´í„°ë¡œ í†µê³„ë¥¼ ê³„ì‚°í•˜ëŠ” ë°©ì‹ì…ë‹ˆë‹¤.
> 
> - **`rolling()`**: ê³ ì •ëœ í¬ê¸°ì˜ ì°½ì„ ì‚¬ìš©í•©ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´ '7ì¼ ì´ë™ í‰ê· 'ì€ ìµœê·¼ 7ì¼ê°„ì˜ ë°ì´í„° í‰ê· ì„ ê³„ì† ê³„ì‚°í•˜ì—¬ ë°ì´í„°ì˜ ë‹¨ê¸° ë³€ë™ì„±ì„ ì™„í™”í•˜ê³  ì¶”ì„¸ë¥¼ ë¶€ë“œëŸ½ê²Œ ë³´ì—¬ì¤ë‹ˆë‹¤.
> - **`expanding()`**: ì°½ì˜ í¬ê¸°ê°€ ë°ì´í„°ê°€ ì¶”ê°€ë ìˆ˜ë¡ ê³„ì† ì»¤ì§‘ë‹ˆë‹¤. ì‹œì‘ì ë¶€í„° í˜„ì¬ ì§€ì ê¹Œì§€ì˜ ëª¨ë“  ë°ì´í„°ë¥¼ ì‚¬ìš©í•˜ì—¬ ëˆ„ì  í•©ê³„, ëˆ„ì  í‰ê·  ë“±ì„ ê³„ì‚°í•  ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.

### `rolling`: ì´ë™ í‰ê· ìœ¼ë¡œ ì¶”ì„¸ íŒŒì•…í•˜ê¸°
> `rolling(window=í¬ê¸°)`ìœ¼ë¡œ ì´ë™ì°½ ê°ì²´ë¥¼ ë§Œë“¤ê³ , `.mean()`, `.std()` ë“± ì§‘ê³„ í•¨ìˆ˜ë¥¼ ì ìš©í•©ë‹ˆë‹¤. `window`ëŠ” ê³„ì‚°ì— í¬í•¨í•  ë°ì´í„°ì˜ ê°œìˆ˜ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.

```python
# ìœ„ì—ì„œ ì‚¬ìš©í•œ ts_df ë°ì´í„°ë¥¼ ê³„ì† ì‚¬ìš©í•©ë‹ˆë‹¤.

# 7ì¼ ì´ë™ í‰ê·  ê³„ì‚°
ts_df['sales_MA_7'] = ts_df['sales'].rolling(window=7).mean()

# 30ì¼ ì´ë™ í‰ê·  ê³„ì‚°
ts_df['sales_MA_30'] = ts_df['sales'].rolling(window=30).mean()

print(ts_df.head(10)) # ì²˜ìŒ 6ê°œëŠ” NaNê°’ì´ ë‚˜ì˜µë‹ˆë‹¤.

# Plotlyë¥¼ ì‚¬ìš©í•˜ì—¬ ì›ë³¸ ë°ì´í„°ì™€ ì´ë™ í‰ê· ì„  ì‹œê°í™”
fig = px.line(ts_df, y=['sales', 'sales_MA_7', 'sales_MA_30'], title='ì¼ë³„ íŒë§¤ëŸ‰ê³¼ ì´ë™ í‰ê· ')
fig.show()
```

### `rolling.apply`:
 > `rolling(window=í¬ê¸°)`ë¡œ ì´ë™ì°½ ê°ì²´ë¥¼ ë§Œë“  ë’¤, `.apply(ì‚¬ìš©ìì •ì˜í•¨ìˆ˜)`ë¥¼ ì‚¬ìš©í•˜ë©´ ì°½(window) ë‚´ì˜ ë°ì´í„°ì— ëŒ€í•´ ì§ì ‘ ì •ì˜í•œ í•¨ìˆ˜ë¥¼ ì ìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
 > ì˜ˆë¥¼ ë“¤ì–´, ì´ë™ì°½ ë‚´ì—ì„œ ì¤‘ì•™ê°’, ìµœëŒ“ê°’, ë˜ëŠ” ë³µì¡í•œ ê³„ì‚°ì„ í•˜ê³  ì‹¶ì„ ë•Œ í™œìš©í•©ë‹ˆë‹¤.


```python
ts_df['sales_Median_7'] = ts_df['sales'].rolling(window=7).apply(np.median)
```

### âœï¸ ì—°ìŠµ ë¬¸ì œ (Practice Problems)

1. `ts_df` ë°ì´í„°ì˜ 'sales' ì»¬ëŸ¼ì— ëŒ€í•´ **14ì¼ ì´ë™ í‘œì¤€í¸ì°¨**ë¥¼ ê³„ì‚°í•˜ì—¬ `sales_MV_14` ë¼ëŠ” ìƒˆ ì»¬ëŸ¼ì— ì €ì¥í•˜ê³ , ìƒìœ„ 20ê°œ í–‰ì„ ì¶œë ¥í•´ë³´ì„¸ìš”. (ì´ë™ í‘œì¤€í¸ì°¨ëŠ” ì£¼ê°€ì˜ ë³€ë™ì„±ì„ ë¶„ì„í•˜ëŠ” ë“± ë°ì´í„°ì˜ ì•ˆì •ì„±ì„ ë³¼ ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.)

```python
# ì—°ìŠµ ë¬¸ì œ 1ë²ˆ í’€ì´ ê³µê°„
ts_df['sales_MV_14'] = ts_df['sales'].rolling(window=14).std()
print(ts_df.head(20))
```

> - **`shift()`**: ë°ì´í„°ì˜ ì¸ë±ìŠ¤ëŠ” ê·¸ëŒ€ë¡œ ë‘” ì±„, ê°’ë§Œ íŠ¹ì • ê¸°ê°„ë§Œí¼ ë’¤ë¡œ(ë˜ëŠ” ì•ìœ¼ë¡œ) ë°€ì–´ëƒ…ë‹ˆë‹¤. ì–´ì œ ë°ì´í„°ì™€ ì˜¤ëŠ˜ ë°ì´í„°ë¥¼ ë‚˜ë€íˆ ë†“ê³  ë¹„êµí•˜ê±°ë‚˜, ì‹œê³„ì—´ ì˜ˆì¸¡ ëª¨ë¸ì˜ í”¼ì²˜(feature)ë¥¼ ë§Œë“¤ ë•Œ ì‚¬ìš©ë©ë‹ˆë‹¤.
> - **`diff()`**: í˜„ì¬ ì‹œì ì˜ ë°ì´í„°ì—ì„œ ë°”ë¡œ ì´ì „ ì‹œì ì˜ ë°ì´í„°ë¥¼ ëº€ ê°’(ì°¨ë¶„)ì„ ê³„ì‚°í•©ë‹ˆë‹¤. ë°ì´í„°ì˜ ì¶”ì„¸(trend)ë¥¼ ì œê±°í•˜ê³  ì •ìƒì„±(stationarity)ì„ ê°€ì§„ ë°ì´í„°ë¡œ ë³€í™˜í•˜ê¸° ìœ„í•´ ì‚¬ìš©ë˜ë©°, ë³€í™”ëŸ‰ ìì²´ë¥¼ ë¶„ì„í•  ë•Œ ìœ ìš©í•©ë‹ˆë‹¤.

```python
# shift() ì˜ˆì œ
# 'sales'ë¥¼ í•œ ì¹¸ ë’¤ë¡œ ë°€ì–´ 'yesterday_sales' ì»¬ëŸ¼ ìƒì„±
ts_df['yesterday_sales'] = ts_df['sales'].shift(1)

# diff() ì˜ˆì œ
# ì–´ì œ ëŒ€ë¹„ ì˜¤ëŠ˜ì˜ íŒë§¤ëŸ‰ ë³€í™” ê³„ì‚°
ts_df['sales_diff'] = ts_df['sales'].diff(1)

print(ts_df[['sales', 'yesterday_sales', 'sales_diff']].head())

# ì¼ë³„ íŒë§¤ëŸ‰ ë³€í™”ëŸ‰ ì‹œê°í™”
fig = px.bar(ts_df, y='sales_diff', title='ì „ì¼ ëŒ€ë¹„ íŒë§¤ëŸ‰ ë³€í™”')
fig.show()
```

### âœï¸ ì—°ìŠµ ë¬¸ì œ (Practice Problems)

1. `ts_df` ë°ì´í„°ì—ì„œ **ì „ì¼ ëŒ€ë¹„ íŒë§¤ëŸ‰ ì¦ê°ë¥ (%)**ì„ ê³„ì‚°í•˜ì—¬ `sales_growth_rate` ì»¬ëŸ¼ì— ì €ì¥í•˜ê³ , ìƒìœ„ 5ê°œ í–‰ì„ ì¶œë ¥í•˜ì„¸ìš”. 
   (ê³µì‹: `(ì˜¤ëŠ˜ íŒë§¤ëŸ‰ - ì–´ì œ íŒë§¤ëŸ‰) / ì–´ì œ íŒë§¤ëŸ‰ * 100`)

```python
# ì—°ìŠµ ë¬¸ì œ 1ë²ˆ í’€ì´ ê³µê°„
ts_df['sales_growth_rate'] = (ts_df['sales'].diff(1) / ts_df['sales'].shift(1)) * 100
# ë˜ëŠ”
# ts_df['sales_growth_rate'] = ts_df['sales'].pct_change() * 100

print(ts_df[['sales', 'sales_growth_rate']].head())
```

## 7. âœï¸ ë¯¸ë‹ˆ ì‹¤ìŠµ PJT ê³¼ì œ

### ê°€ìƒ ì£¼ê°€ ë°ì´í„° ë¶„ì„

**ì‹œë‚˜ë¦¬ì˜¤**: ë‹¹ì‹ ì€ ë°ì´í„° ë¶„ì„ê°€ì…ë‹ˆë‹¤. í•œ íšŒì‚¬ì˜ ì•½ 1ë…„ê°„ì˜ ì£¼ê°€ ë°ì´í„°ë¥¼ ë°›ì•„ ë¶„ì„ ë¦¬í¬íŠ¸ë¥¼ ì‘ì„±í•´ì•¼ í•©ë‹ˆë‹¤. Pandas ì‹œê³„ì—´ ê¸°ëŠ¥ì„ í™œìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ë¶„ì„í•˜ê³ , ì£¼ìš” ì¸ì‚¬ì´íŠ¸ë¥¼ ì‹œê°í™”í•˜ì—¬ ë³´ê³ í•˜ì„¸ìš”.

**ë°ì´í„°ì…‹**: ê°€ìƒì˜ ì¼ë³„ ì£¼ê°€ ë°ì´í„° (Open, High, Low, Close, Volume)

**ë¶„ì„ ëª©í‘œ**: 
1. ì›”ë³„ í‰ê·  ì¢…ê°€(Close)ë¥¼ ê³„ì‚°í•˜ê³  ì‹œê°í™”í•˜ì—¬ ì „ë°˜ì ì¸ ì£¼ê°€ ì¶”ì„¸ë¥¼ íŒŒì•…í•©ë‹ˆë‹¤.
2. 20ì¼ ì´ë™ í‰ê· ì„ ê³¼ 60ì¼ ì´ë™ í‰ê· ì„ ì„ ê³„ì‚°í•˜ê³ , ì›ë³¸ ì¢…ê°€ ë°ì´í„°ì™€ í•¨ê»˜ ì‹œê°í™”í•˜ì—¬ ë‹¨ê¸° ë° ì¤‘ê¸° ì¶”ì„¸ë¥¼ ë¶„ì„í•©ë‹ˆë‹¤.
3. ì¼ë³„ ìˆ˜ìµë¥ (daily return)ì„ ê³„ì‚°í•˜ê³ , ë¶„í¬ë¥¼ íˆìŠ¤í† ê·¸ë¨ìœ¼ë¡œ ì‹œê°í™”í•˜ì—¬ ì£¼ê°€ì˜ ë³€ë™ì„±ì„ í™•ì¸í•©ë‹ˆë‹¤.

```python
# 1. ê°€ìƒ ì£¼ê°€ ë°ì´í„° ìƒì„±
np.random.seed(101)
dates = pd.date_range('2022-01-01', '2022-12-31', freq='B') # Business day
price_changes = np.random.randn(len(dates)).cumsum()
start_price = 100
close_prices = start_price + price_changes

stock_df = pd.DataFrame({'Close': close_prices}, index=dates)
stock_df['Open'] = stock_df['Close'].shift(1) + np.random.uniform(-2, 2, len(stock_df))
stock_df['High'] = stock_df[['Open', 'Close']].max(axis=1) + np.random.uniform(0, 3, len(stock_df))
stock_df['Low'] = stock_df[['Open', 'Close']].min(axis=1) - np.random.uniform(0, 3, len(stock_df))
stock_df['Volume'] = np.random.randint(100000, 500000, len(stock_df))
stock_df = stock_df.dropna()

print("--- ìƒì„±ëœ ì£¼ê°€ ë°ì´í„° ---")
print(stock_df.head())
```

### **ë¶„ì„ ê³¼ì œ 1: ì›”ë³„ í‰ê·  ì¢…ê°€ ê³„ì‚° ë° ì‹œê°í™”**

ğŸ’¡ HINT : ì›”ë³„ í‰ê·  ì¢…ê°€ë¥¼ ê³„ì‚°í•˜ë ¤ë©´ `resample()` ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ë°ì´í„°ë¥¼ ì›”ë³„ë¡œ ì§‘ê³„í•˜ê³ , `mean()`ì„ ì ìš©í•˜ì„¸ìš”. ì‹œê°í™”ëŠ” `plotly.express.line()`ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.

```python
# ì›”ë³„ í‰ê·  ì¢…ê°€ ê³„ì‚° (ë¦¬ìƒ˜í”Œë§)
monthly_avg_close = stock_df['Close'].resample('MS').mean()

print("--- ì›”ë³„ í‰ê·  ì¢…ê°€ ---")
print(monthly_avg_close)

# ì‹œê°í™”
fig1 = px.line(monthly_avg_close, title='ì›”ë³„ í‰ê·  ì¢…ê°€ ì¶”ì´', markers=True)
fig1.update_traces(name='ì›”í‰ê· ', showlegend=True)
fig1.show()
```

### **ë¶„ì„ ê³¼ì œ 2: ì´ë™ í‰ê· ì„  ê³„ì‚° ë° ì‹œê°í™”**
ğŸ’¡ HINT : ì´ë™ í‰ê· ì„ ì„ ê³„ì‚°í•˜ë ¤ë©´ `rolling()` ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ì´ë™ í‰ê· ì„ ê³„ì‚°í•˜ê³ , `mean()`ì„ ì ìš©í•˜ì„¸ìš”. ì‹œê°í™”ëŠ” `plotly.express.line()`ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.


```python
# 20ì¼, 60ì¼ ì´ë™ í‰ê·  ê³„ì‚° (ë¡¤ë§)
stock_df['MA20'] = stock_df['Close'].rolling(window=20).mean()
stock_df['MA60'] = stock_df['Close'].rolling(window=60).mean()

# ì‹œê°í™”
fig2 = px.line(stock_df, y=['Close', 'MA20', 'MA60'], title='ì¢…ê°€ ë° ì´ë™ í‰ê· ì„ ')
fig2.show()
```

### **ë¶„ì„ ê³¼ì œ 3: ì¼ë³„ ìˆ˜ìµë¥  ê³„ì‚° ë° ì‹œê°í™”**
ğŸ’¡ HINT : ì¼ë³„ ìˆ˜ìµë¥ ì€ `pct_change()` ë©”ì„œë“œë¥¼ ì‚¬ìš©í•˜ì—¬ ê³„ì‚°í•˜ê³ , ì‹œê°í™”ëŠ” `plotly.express.histogram()`ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.
