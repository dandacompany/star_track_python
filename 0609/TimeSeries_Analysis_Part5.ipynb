## 1\. ì‹œê³„ì—´ ë°ì´í„°ì˜ ì´í•´ì™€ ì •ìƒì„± (Understanding Time Series & Stationarity)

### ğŸ’¡ ê°œë… (Concept)

\*\*ì‹œê³„ì—´ ë°ì´í„°(Time Series Data)\*\*ë€ ì¼ì • ì‹œê°„ ê°„ê²©ìœ¼ë¡œ ì¸¡ì •ëœ ë°ì´í„°ì˜ ìˆœì°¨ì ì¸ ì§‘í•©ì„ ì˜ë¯¸í•©ë‹ˆë‹¤. ì£¼ì‹ ê°€ê²©, ì›”ë³„ ê°•ìˆ˜ëŸ‰, ì¼ì¼ ì›¹ì‚¬ì´íŠ¸ ë°©ë¬¸ì ìˆ˜ ë“±ì´ ëª¨ë‘ ì‹œê³„ì—´ ë°ì´í„°ì˜ ì˜ˆì…ë‹ˆë‹¤.

ì‹œê³„ì—´ ë¶„ì„ì˜ í•µì‹¬ ëª©í‘œëŠ” ë°ì´í„°ì— ìˆ¨ê²¨ì§„ íŒ¨í„´ì„ ë°œê²¬í•˜ê³  ì´ë¥¼ ë°”íƒ•ìœ¼ë¡œ ë¯¸ë˜ ê°’ì„ ì˜ˆì¸¡í•˜ëŠ” ê²ƒì…ë‹ˆë‹¤. ëŒ€ë¶€ë¶„ì˜ ì‹œê³„ì—´ ì˜ˆì¸¡ ëª¨ë¸ì€ ë°ì´í„°ê°€ \*\*ì •ìƒì„±(Stationarity)\*\*ì„ ê°–ëŠ”ë‹¤ê³  ê°€ì •í•©ë‹ˆë‹¤.

**ì •ìƒì„±ì´ë€?**
ì‹œê°„ì´ í˜ëŸ¬ë„ ë°ì´í„°ì˜ í†µê³„ì  íŠ¹ì„±(í‰ê· , ë¶„ì‚° ë“±)ì´ ë³€í•˜ì§€ ì•ŠëŠ” ìƒíƒœë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤. ì¦‰, ì‹œê³„ì—´ ë°ì´í„°ì˜ ë¶„í¬ê°€ ì‹œê°„ì— ë”°ë¼ ì¼ì •í•´ì•¼ í•©ë‹ˆë‹¤.

  * **ì™œ ì •ìƒì„±ì´ ì¤‘ìš”í•œê°€?**: ë°ì´í„°ì˜ íŒ¨í„´ì´ ì‹œê°„ê³¼ ê´€ê³„ì—†ì´ ì¼ì •í•´ì•¼ ê³¼ê±°ì˜ íŒ¨í„´ì„ ê¸°ë°˜ìœ¼ë¡œ ë¯¸ë˜ë¥¼ ì•ˆì •ì ìœ¼ë¡œ ì˜ˆì¸¡í•  ìˆ˜ ìˆê¸° ë•Œë¬¸ì…ë‹ˆë‹¤.
  * **ë¹„ì •ìƒ ë°ì´í„°**: ì£¼ê°€ì²˜ëŸ¼ ì§€ì†ì ìœ¼ë¡œ ìƒìŠ¹í•˜ëŠ” ì¶”ì„¸(Trend)ê°€ ìˆê±°ë‚˜, ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰ì²˜ëŸ¼ ê³„ì ˆì— ë”°ë¼ ì£¼ê¸°ì ìœ¼ë¡œ ë³€ë™í•˜ëŠ” ê³„ì ˆì„±(Seasonality)ì´ ìˆëŠ” ë°ì´í„°ëŠ” ë¹„ì •ìƒ ì‹œê³„ì—´ì…ë‹ˆë‹¤.

**ì •ìƒì„± í™•ì¸ ë°©ë²•: ADF ê²€ì • (Augmented Dickey-Fuller Test)**
ADF ê²€ì •ì€ ì‹œê³„ì—´ ë°ì´í„°ì˜ ì •ìƒì„±ì„ í†µê³„ì ìœ¼ë¡œ í™•ì¸í•˜ëŠ” ëŒ€í‘œì ì¸ ë°©ë²•ì…ë‹ˆë‹¤. ê°€ì„¤ì€ ë‹¤ìŒê³¼ ê°™ìŠµë‹ˆë‹¤.

  * **ê·€ë¬´ê°€ì„¤(Null Hypothesis)**: ë°ì´í„°ì— ë‹¨ìœ„ê·¼(unit root)ì´ ì¡´ì¬í•œë‹¤. (ì¦‰, ë¹„ì •ìƒ ì‹œê³„ì—´ì´ë‹¤)
  * **ëŒ€ë¦½ê°€ì„¤(Alternative Hypothesis)**: ë°ì´í„°ê°€ ì •ìƒì„±ì„ ë§Œì¡±í•œë‹¤.

ê²€ì • ê²°ê³¼ë¡œ ë‚˜ì˜¤ëŠ” **p-value**ê°€ íŠ¹ì • ìœ ì˜ìˆ˜ì¤€(ë³´í†µ 0.05)ë³´ë‹¤ ì‘ìœ¼ë©´ ê·€ë¬´ê°€ì„¤ì„ ê¸°ê°í•˜ê³ , ë°ì´í„°ê°€ ì •ìƒì„±ì„ ê°–ëŠ”ë‹¤ê³  íŒë‹¨í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.

**ë¹„ì •ìƒ ë°ì´í„°ë¥¼ ì •ìƒ ë°ì´í„°ë¡œ ë³€í™˜í•˜ê¸°: ì°¨ë¶„ (Differencing)**
ëŒ€ë¶€ë¶„ì˜ ë¹„ì •ìƒ ì‹œê³„ì—´ì€ **ì°¨ë¶„**ì„ í†µí•´ ì •ìƒì„±ì„ ë§Œì¡±í•˜ëŠ” ë°ì´í„°ë¡œ ë³€í™˜í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤. ì°¨ë¶„ì€ í˜„ì¬ ì‹œì ì˜ ë°ì´í„°ì—ì„œ ë°”ë¡œ ì´ì „ ì‹œì ì˜ ë°ì´í„°ë¥¼ ë¹¼ì£¼ëŠ” ê°„ë‹¨í•œ ê³¼ì •ì…ë‹ˆë‹¤. ì´ë¥¼ í†µí•´ ë°ì´í„°ì˜ ì¶”ì„¸ë¥¼ ì œê±°í•˜ëŠ” íš¨ê³¼ë¥¼ ì–»ì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

$$Y'_t = Y_t - Y_{t-1}$$

### ğŸ’» ì˜ˆì‹œ ì½”ë“œ (Example Code)

í•­ê³µ ìŠ¹ê° ìˆ˜ ë°ì´í„°ë¥¼ í™œìš©í•˜ì—¬ ì‹œê³„ì—´ ë°ì´í„°ì˜ íŠ¹ì§•ì„ ì‹œê°í™”í•˜ê³ , ADF ê²€ì •ê³¼ ì°¨ë¶„ì„ ì‹¤ìŠµí•´ ë³´ê² ìŠµë‹ˆë‹¤.

```
import pandas as pd
import plotly.express as px
from statsmodels.tsa.stattools import adfuller

# 1. ë°ì´í„° ë¡œë“œ ë° ì‹œê°í™”
# ì›¹ì—ì„œ ë°”ë¡œ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url, index_col='Month', parse_dates=True)
df.columns = ['Passengers']

# ì‹œê³„ì—´ ë°ì´í„° í”Œë¡¯
fig = px.line(df, x=df.index, y='Passengers', title='ì›”ë³„ í•­ê³µ ìŠ¹ê° ìˆ˜ (1949-1960)')
fig.show()

# 2. ADF ê²€ì • í•¨ìˆ˜ ì •ì˜ ë° ì •ìƒì„± í™•ì¸
def adf_test(series, name=''):
    """ì£¼ì–´ì§„ ì‹œê³„ì—´ì— ëŒ€í•´ ADF ê²€ì •ì„ ìˆ˜í–‰í•˜ê³  ê²°ê³¼ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤."""
    print(f'>>> {name} ADF Test <<<')
    result = adfuller(series.dropna()) # NaN ê°’ì„ ì œì™¸í•˜ê³  ê²€ì • ìˆ˜í–‰
    labels = ['ADF Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used']
    for value, label in zip(result, labels):
        print(label + ' : ' + str(value))

    if result[1] <= 0.05:
        print("=> p-valueê°€ 0.05ë³´ë‹¤ ì‘ìœ¼ë¯€ë¡œ, ê·€ë¬´ê°€ì„¤ì„ ê¸°ê°í•©ë‹ˆë‹¤. ë°ì´í„°ëŠ” ì •ìƒì„±ì„ ë§Œì¡±í•©ë‹ˆë‹¤.\n")
    else:
        print("=> p-valueê°€ 0.05ë³´ë‹¤ í¬ë¯€ë¡œ, ê·€ë¬´ê°€ì„¤ì„ ê¸°ê°í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤. ë°ì´í„°ëŠ” ë¹„ì •ìƒ ì‹œê³„ì—´ì…ë‹ˆë‹¤.\n")

# ì›ë³¸ ë°ì´í„°ì— ëŒ€í•œ ADF ê²€ì •
adf_test(df['Passengers'], name='Original')

# 3. 1ì°¨ ì°¨ë¶„ì„ í†µí•œ ì •ìƒì„± í™•ë³´
df['Passengers_diff'] = df['Passengers'].diff()

# 1ì°¨ ì°¨ë¶„ ë°ì´í„° í”Œë¡¯
fig_diff = px.line(df, x=df.index, y='Passengers_diff', title='1ì°¨ ì°¨ë¶„ëœ í•­ê³µ ìŠ¹ê° ìˆ˜')
fig_diff.show()

# 1ì°¨ ì°¨ë¶„ ë°ì´í„°ì— ëŒ€í•œ ADF ê²€ì •
adf_test(df['Passengers_diff'], name='1st Differenced')
```

## 2\. ë‹¨ë³€ëŸ‰ ì‹œê³„ì—´ ë¶„ì„ (Univariate Analysis): ARIMA

### ğŸ’¡ ê°œë… (Concept)

**ARIMA(AutoRegressive Integrated Moving Average)** ëª¨ë¸ì€ ê°€ì¥ ëŒ€í‘œì ì¸ ë‹¨ë³€ëŸ‰ ì‹œê³„ì—´ ì˜ˆì¸¡ ëª¨ë¸ì…ë‹ˆë‹¤. ì¦‰, í•˜ë‚˜ì˜ ë³€ìˆ˜(y)ì˜ ê³¼ê±° ë°ì´í„° íŒ¨í„´ê³¼ ì˜¤ì°¨ë¥¼ í•™ìŠµí•˜ì—¬ ë¯¸ë˜ ê°’ì„ ì˜ˆì¸¡í•©ë‹ˆë‹¤. ARIMAëŠ” ì„¸ ê°€ì§€ í•µì‹¬ ìš”ì†Œë¡œ êµ¬ì„±ë©ë‹ˆë‹¤.

  * **AR (AutoRegressive, ìê¸°íšŒê·€)**: **ê³¼ê±°ì˜ ê°’**ì´ í˜„ì¬ì˜ ê°’ì— ì˜í–¥ì„ ë¯¸ì¹œë‹¤ëŠ” ì•„ì´ë””ì–´ì— ê¸°ë°˜í•©ë‹ˆë‹¤.

      * `p`ëŠ” ì°¸ê³ í•  ê³¼ê±° ì‹œì ì˜ ê°œìˆ˜ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤. (ì˜ˆ: `p=3`ì´ë©´ 3ê°œì›” ì „ ë°ì´í„°ê¹Œì§€ ì°¸ê³ )
      * ë©”íƒ€í¬: "ì´ë²ˆ ë‹¬ ë§¤ì¶œì€ ì§€ë‚œë‹¬ê³¼ ì§€ì§€ë‚œë‹¬ ë§¤ì¶œê³¼ ë¹„ìŠ·í•  ê±°ì•¼."
        $$Y_t = \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} + \epsilon_t$$

  * **I (Integrated, í†µí•©)**: ë¹„ì •ìƒ ì‹œê³„ì—´ì„ ì •ìƒ ì‹œê³„ì—´ë¡œ ë§Œë“¤ê¸° ìœ„í•´ í•„ìš”í•œ **ì°¨ë¶„ íšŸìˆ˜**ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.

      * `d`ëŠ” ì°¨ë¶„ì„ ì ìš©í•œ íšŸìˆ˜ë¥¼ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤. (d=1ì´ë©´ 1ì°¨ ì°¨ë¶„)

  * **MA (Moving Average, ì´ë™í‰ê· )**: **ê³¼ê±°ì˜ ì˜ˆì¸¡ ì˜¤ì°¨**ê°€ í˜„ì¬ì˜ ê°’ì— ì˜í–¥ì„ ë¯¸ì¹œë‹¤ëŠ” ê°œë…ì…ë‹ˆë‹¤.

      * `q`ëŠ” ì°¸ê³ í•  ê³¼ê±° ì˜ˆì¸¡ ì˜¤ì°¨ì˜ ê°œìˆ˜ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.
      * ë©”íƒ€í¬: "ì§€ë‚œë‹¬ ë§¤ì¶œ ì˜ˆì¸¡ì´ ë¹—ë‚˜ê°”ìœ¼ë‹ˆ, ê·¸ ì˜¤ì°¨ë¥¼ ë°˜ì˜í•´ì„œ ì´ë²ˆ ë‹¬ ì˜ˆì¸¡ì„ ë³´ì •í•´ì•¼ í•´."
        $$Y_t = \epsilon_t + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \cdots + \theta_q \epsilon_{t-q}$$

**ACFì™€ PACF: pì™€ q ì •í•˜ê¸°**
ARIMA ëª¨ë¸ì˜ ì°¨ìˆ˜ `p`ì™€ `q`ë¥¼ ê²°ì •í•˜ê¸° ìœ„í•´ \*\*ACF(ìê¸°ìƒê´€í•¨ìˆ˜)\*\*ì™€ **PACF(ë¶€ë¶„ìê¸°ìƒê´€í•¨ìˆ˜)** í”Œë¡¯ì„ í™œìš©í•©ë‹ˆë‹¤.

  * **ACF (Autocorrelation Function)**: ì‹œì  tì˜ ë°ì´í„°ì™€ ê·¸ ì´ì „ ì‹œì ë“¤(t-1, t-2, ...)ì˜ ë°ì´í„° ê°„ì˜ ìƒê´€ê´€ê³„ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤. **MA ëª¨ë¸ì˜ ì°¨ìˆ˜ që¥¼ ê²°ì •**í•˜ëŠ” ë° ì‚¬ìš©ë©ë‹ˆë‹¤.
  * **PACF (Partial Autocorrelation Function)**: ë‹¤ë¥¸ ì‹œì ë“¤ì˜ ì˜í–¥ì„ ì œì™¸í•˜ê³ , ì˜¤ì§ ë‘ ì‹œì (tì™€ t-k)ì˜ ë°ì´í„° ê°„ ìˆœìˆ˜í•œ ìƒê´€ê´€ê³„ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤. **AR ëª¨ë¸ì˜ ì°¨ìˆ˜ pë¥¼ ê²°ì •**í•˜ëŠ” ë° ì‚¬ìš©ë©ë‹ˆë‹¤.

| ëª¨ë¸ | ACF í”Œë¡¯ íŠ¹ì§• | PACF í”Œë¡¯ íŠ¹ì§• |
| :--- | :--- | :--- |
| **AR(p)** | ì ì°¨ ê°ì†Œí•˜ë©° 0ì— ìˆ˜ë ´ | `p` ì‹œì  ì´í›„ ê¸‰ê²©íˆ 0ìœ¼ë¡œ ì ˆë‹¨ |
| **MA(q)** | `q` ì‹œì  ì´í›„ ê¸‰ê²©íˆ 0ìœ¼ë¡œ ì ˆë‹¨ | ì ì°¨ ê°ì†Œí•˜ë©° 0ì— ìˆ˜ë ´ |

### ğŸ’» ì˜ˆì‹œ ì½”ë“œ (Example Code)

ì´ë²ˆì—ëŠ” `moneystock-spending-series.csv` ë°ì´í„°ì…‹ì„ ì‚¬ìš©í•˜ì—¬ ARIMA ëª¨ë¸ì„ êµ¬ì¶•í•˜ê³  ì˜ˆì¸¡ì„ ìˆ˜í–‰í•©ë‹ˆë‹¤. ì´ ë°ì´í„°ëŠ” M2 í†µí™”ëŸ‰ê³¼ ê°œì¸ ì†Œë¹„ ì§€ì¶œ ë°ì´í„°ë¡œ êµ¬ì„±ë˜ì–´ ìˆìŠµë‹ˆë‹¤.

```
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# ë°ì´í„° ë¡œë“œ (money stock spending)
df_money = pd.read_csv('https://raw.githubusercontent.com/datamanim/datarepo/main/tsa/moneystock-spending-series.csv', index_col=0, parse_dates=True)
df_money.index.freq = 'MS' # ì›”ë³„ ë°ì´í„° ì£¼íŒŒìˆ˜ ì„¤ì •

# 'Money' ë³€ìˆ˜ì— ëŒ€í•´ ë¶„ì„ ì§„í–‰
money_series = df_money['Money']

# 1. ì •ìƒì„± í™•ì¸ ë° ì°¨ë¶„
# ADF ê²€ì • ê²°ê³¼ p-valueê°€ 1.0ìœ¼ë¡œ ë¹„ì •ìƒ ì‹œê³„ì—´ì„ (ìœ„ì—ì„œ ì‹¤ìŠµ)
# 2ì°¨ ì°¨ë¶„ì„ í†µí•´ ì •ìƒì„± í™•ë³´
money_diff2 = money_series.diff().diff().dropna()
# adf_test(money_diff2) # p-valueê°€ ë§¤ìš° ì‘ê²Œ ë‚˜ì™€ ì •ìƒì„± ë§Œì¡±

# 2. ACF, PACF í”Œë¡¯ìœ¼ë¡œ p, q ì¶”ì •
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
plot_acf(money_diff2, ax=ax1)
plot_pacf(money_diff2, ax=ax2)
plt.show()
# PACF í”Œë¡¯ì„ ë³´ë©´ lag 1 ì´í›„ì— ê°’ì´ ê¸‰ê²©íˆ ê°ì†Œí•˜ë¯€ë¡œ p=1ë¡œ ì¶”ì • ê°€ëŠ¥
# ACF í”Œë¡¯ì„ ë³´ë©´ lag 1 ì´í›„ì— ê°’ì´ ê¸‰ê²©íˆ ê°ì†Œí•˜ë¯€ë¡œ q=1ë¡œ ì¶”ì • ê°€ëŠ¥

# 3. ARIMA(1, 2, 1) ëª¨ë¸ êµ¬ì¶• ë° í•™ìŠµ
# p=1, d=2(2ì°¨ ì°¨ë¶„), q=1
model = ARIMA(money_series, order=(1, 2, 1))
fitted_model = model.fit()
print(fitted_model.summary())

# 4. ì˜ˆì¸¡ ë° ì‹œê°í™”
# í•™ìŠµ ë°ì´í„° ë¶„í•  (ë§ˆì§€ë§‰ 12ê°œì›”ì„ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ)
train_data = money_series[:-12]
test_data = money_series[-12:]

# ARIMA(1, 2, 1) ëª¨ë¸ ì¬í•™ìŠµ
model_train = ARIMA(train_data, order=(1, 2, 1))
fitted_train_model = model_train.fit()

# ì˜ˆì¸¡ (12ê°œì›”)
forecast = fitted_train_model.forecast(steps=12)

# ê²°ê³¼ ì‹œê°í™”
fig = go.Figure()
fig.add_trace(go.Scatter(x=train_data.index, y=train_data, name='Train'))
fig.add_trace(go.Scatter(x=test_data.index, y=test_data, name='Test'))
fig.add_trace(go.Scatter(x=forecast.index, y=forecast, name='Forecast'))
fig.update_layout(title='M2 í†µí™”ëŸ‰: ì‹¤ì œê°’ vs ì˜ˆì¸¡ê°’ (ARIMA)', xaxis_title='ë‚ ì§œ', yaxis_title='í†µí™”ëŸ‰')
fig.show()
```

## 3\. ê³„ì ˆì„±ì„ ê³ ë ¤í•œ ì‹œê³„ì—´ ë¶„ì„ (Seasonal Analysis): SARIMA

### ğŸ’¡ ê°œë… (Concept)

í˜„ì‹¤ì˜ ë§ì€ ì‹œê³„ì—´ ë°ì´í„°ëŠ” **ê³„ì ˆì„±(Seasonality)**, ì¦‰ íŠ¹ì • ê¸°ê°„(ì£¼, ì›”, ë¶„ê¸°, ë…„ ë“±)ì„ ì£¼ê¸°ë¡œ ë°˜ë³µë˜ëŠ” íŒ¨í„´ì„ ê°€ì§‘ë‹ˆë‹¤. **SARIMA(Seasonal ARIMA)** ëª¨ë¸ì€ ê¸°ì¡´ ARIMA ëª¨ë¸ì— ì´ëŸ¬í•œ ê³„ì ˆì  ìš”ì¸ì„ ì¶”ê°€í•˜ì—¬ ì˜ˆì¸¡ ì„±ëŠ¥ì„ ë†’ì¸ ëª¨ë¸ì…ë‹ˆë‹¤.

SARIMA ëª¨ë¸ì€ `ARIMA(p,d,q)`ì— ê³„ì ˆì„± ìš”ì†Œë¥¼ ë‚˜íƒ€ë‚´ëŠ” `(P,D,Q,s)`ê°€ ì¶”ê°€ë©ë‹ˆë‹¤.

  * **`s`**: ê³„ì ˆ ì£¼ê¸°ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤. (ì˜ˆ: ì›”ë³„ ë°ì´í„°ë©´ `s=12`, ë¶„ê¸°ë³„ ë°ì´í„°ë©´ `s=4`)
  * **`P`**: ê³„ì ˆì„± ìê¸°íšŒê·€(Seasonal AR) ì°¨ìˆ˜. `s` ê¸°ê°„ ì „ì˜ ê³¼ê±° ê°’ì´ í˜„ì¬ ê°’ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.
  * **`D`**: ê³„ì ˆì„± ì°¨ë¶„(Seasonal Differencing) íšŸìˆ˜. ê³„ì ˆì„± íŒ¨í„´ì„ ì œê±°í•˜ê¸° ìœ„í•´ í•„ìš”í•œ ì°¨ë¶„ íšŸìˆ˜ì…ë‹ˆë‹¤.
  * **`Q`**: ê³„ì ˆì„± ì´ë™í‰ê· (Seasonal MA) ì°¨ìˆ˜. `s` ê¸°ê°„ ì „ì˜ ê³¼ê±° ì˜ˆì¸¡ ì˜¤ì°¨ê°€ í˜„ì¬ ê°’ì— ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ ë‚˜íƒ€ëƒ…ë‹ˆë‹¤.

ë©”íƒ€í¬: "ì˜¬í•´ 8ì›”ì˜ ì•„ì´ìŠ¤í¬ë¦¼ íŒë§¤ëŸ‰ì€ ì‘ë…„ 8ì›” íŒë§¤ëŸ‰ê³¼ ì‘ë…„ ì˜ˆì¸¡ ì˜¤ì°¨ì— ì˜í–¥ì„ ë°›ì„ ê²ƒì´ë‹¤."

**Auto-ARIMA: ìµœì ì˜ íŒŒë¼ë¯¸í„° ìë™ íƒìƒ‰**
SARIMA ëª¨ë¸ì€ `(p,d,q)(P,D,Q,s)`ë¼ëŠ” 7ê°œì˜ íŒŒë¼ë¯¸í„°ë¥¼ ê²°ì •í•´ì•¼ í•˜ë¯€ë¡œ, ACF/PACFë§Œìœ¼ë¡œ ìµœì ì˜ ì¡°í•©ì„ ì°¾ê¸° ì–´ë µìŠµë‹ˆë‹¤. ì´ë•Œ **`auto_arima`** í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•˜ë©´ ë§¤ìš° í¸ë¦¬í•©ë‹ˆë‹¤. `auto_arima`ëŠ” ì£¼ì–´ì§„ ë°ì´í„°ì— ëŒ€í•´ ë‹¤ì–‘í•œ íŒŒë¼ë¯¸í„° ì¡°í•©ì„ ì‹œë„í•˜ê³ , **AIC(Akaike Information Criterion)** ê°’ì´ ê°€ì¥ ë‚®ì€ ìµœì ì˜ ëª¨ë¸ì„ ìë™ìœ¼ë¡œ ì°¾ì•„ì¤ë‹ˆë‹¤.

### ğŸ’» ì˜ˆì‹œ ì½”ë“œ (Example Code)

`auto_arima`ë¥¼ ì‚¬ìš©í•˜ì—¬ ì›”ë³„ í•­ê³µ ìŠ¹ê° ìˆ˜ ë°ì´í„°ì— ëŒ€í•œ ìµœì ì˜ SARIMA ëª¨ë¸ì„ ì°¾ê³ , ì˜ˆì¸¡ì„ ìˆ˜í–‰í•´ ë³´ê² ìŠµë‹ˆë‹¤.

```
# pmdarima ë¼ì´ë¸ŒëŸ¬ë¦¬ê°€ í•„ìš”í•©ë‹ˆë‹¤.
# !pip install pmdarima
import pandas as pd
import plotly.graph_objects as go
import pmdarima as pm
from pmdarima.model_selection import train_test_split

# ë°ì´í„° ë¡œë“œ
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
df = pd.read_csv(url, index_col='Month', parse_dates=True)
df.columns = ['Passengers']

# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶„ë¦¬
train_auto, test_auto = train_test_split(df['Passengers'], train_size=len(df)-12)

# auto_arimaë¡œ ìµœì  ëª¨ë¸ íƒìƒ‰
# ê³„ì ˆ ì£¼ê¸°ëŠ” ì›”ë³„ ë°ì´í„°ì´ë¯€ë¡œ m=12ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
sarima_model = pm.auto_arima(train_auto,
                             start_p=1, start_q=1,
                             test='adf',       # adf ê²€ì •ìœ¼ë¡œ d ê²°ì •
                             max_p=3, max_q=3,  # p, q ìµœëŒ€ê°’
                             m=12,             # ê³„ì ˆ ì£¼ê¸°
                             d=None,           # dëŠ” ìë™ìœ¼ë¡œ ì°¾ë„ë¡ ì„¤ì •
                             seasonal=True,    # ê³„ì ˆì„± ëª¨ë¸ ì‚¬ìš©
                             start_P=0,
                             D=1,              # ê³„ì ˆì„± ì°¨ë¶„ì€ 1ë¡œ ê³ ì • (ì¼ë°˜ì )
                             trace=True,       # ëª¨ë¸ íƒìƒ‰ ê³¼ì • ì¶œë ¥
                             error_action='ignore',
                             suppress_warnings=True,
                             stepwise=True)     # ìµœì  íŒŒë¼ë¯¸í„°ë¥¼ íš¨ìœ¨ì ìœ¼ë¡œ ì°¾ìŒ

print(sarima_model.summary())

# í…ŒìŠ¤íŠ¸ ë°ì´í„°ì— ëŒ€í•œ ì˜ˆì¸¡
n_periods = 12
forecast_auto, conf_int = sarima_model.predict(n_periods=n_periods, return_conf_int=True)

# ì˜ˆì¸¡ ê²°ê³¼ ì‹œê°í™”
forecast_index = pd.date_range(start=train_auto.index[-1] + pd.DateOffset(months=1), periods=n_periods, freq='MS')

fig = go.Figure()
fig.add_trace(go.Scatter(x=df.index, y=df['Passengers'], name='Actual'))
fig.add_trace(go.Scatter(x=forecast_index, y=forecast_auto, name='Forecast'))

# ì‹ ë¢°êµ¬ê°„ ì¶”ê°€
fig.add_trace(go.Scatter(x=forecast_index, y=conf_int[:, 0], fill='tonexty',
                         mode='lines', line_color='rgba(255,0,0,0.1)', name='Lower CI'))
fig.add_trace(go.Scatter(x=forecast_index, y=conf_int[:, 1], fill='tonexty',
                         mode='lines', line_color='rgba(255,0,0,0.1)', name='Upper CI'))

fig.update_layout(title='í•­ê³µ ìŠ¹ê° ìˆ˜ ì˜ˆì¸¡ (SARIMA with auto_arima)', xaxis_title='ë‚ ì§œ', yaxis_title='ìŠ¹ê° ìˆ˜')
fig.show()
```

## 4\. ë‹¤ë³€ëŸ‰ ì‹œê³„ì—´ ë¶„ì„ (Multivariate Analysis): VAR

### ğŸ’¡ ê°œë… (Concept)

**VAR(Vector AutoRegression, ë²¡í„° ìê¸°íšŒê·€)** ëª¨ë¸ì€ **ì—¬ëŸ¬ ê°œì˜ ì‹œê³„ì—´ ë³€ìˆ˜**ê°€ ì„œë¡œ ì˜í–¥ì„ ì£¼ê³ ë°›ëŠ” ê´€ê³„ë¥¼ ëª¨ë¸ë§í•˜ëŠ” **ë‹¤ë³€ëŸ‰(Multivariate)** ì‹œê³„ì—´ ë¶„ì„ ê¸°ë²•ì…ë‹ˆë‹¤.

ë‹¨ë³€ëŸ‰ ë¶„ì„(ARIMA ë“±)ì´ í•˜ë‚˜ì˜ ë³€ìˆ˜(y)ì˜ ê³¼ê±°ë§Œì„ ë³´ê³  ë¯¸ë˜ë¥¼ ì˜ˆì¸¡í–ˆë‹¤ë©´, VARì€ ì—¬ëŸ¬ ë³€ìˆ˜(y1, y2, ...)ì˜ ê³¼ê±° ê°’ë“¤ì´ ì„œë¡œì˜ í˜„ì¬ ê°’ì— ì–´ë–»ê²Œ ì˜í–¥ì„ ë¯¸ì¹˜ëŠ”ì§€ë¥¼ ë™ì‹œì— ë¶„ì„í•©ë‹ˆë‹¤.

ë©”íƒ€í¬: "ì˜¤ëŠ˜ì˜ ê¸°ì˜¨(y1)ì€ ì–´ì œì˜ ê¸°ì˜¨ë¿ë§Œ ì•„ë‹ˆë¼ ì–´ì œì˜ í’ì†(y2)ê³¼ ìŠµë„(y3)ì—ë„ ì˜í–¥ì„ ë°›ëŠ”ë‹¤."

**VAR ëª¨ë¸ì˜ íŠ¹ì§•**:

  * **ë‚´ìƒì„±**: ëª¨ë¸ì— í¬í•¨ëœ ëª¨ë“  ë³€ìˆ˜ëŠ” ì„œë¡œì—ê²Œ ì˜í–¥ì„ ì£¼ëŠ” ë‚´ìƒë³€ìˆ˜(endogenous variable)ë¡œ ì·¨ê¸‰ë©ë‹ˆë‹¤.
  * **ë°©ì •ì‹ ì‹œìŠ¤í…œ**: ê° ë³€ìˆ˜ì— ëŒ€í•´ ë³„ë„ì˜ íšŒê·€ ë°©ì •ì‹ì´ ë§Œë“¤ì–´ì§‘ë‹ˆë‹¤. ì˜ˆë¥¼ ë“¤ì–´, ë³€ìˆ˜ê°€ 2ê°œ(y1, y2)ì¸ VAR(1) ëª¨ë¸ì€ ë‹¤ìŒê³¼ ê°™ì€ ë‘ ê°œì˜ ë°©ì •ì‹ìœ¼ë¡œ êµ¬ì„±ë©ë‹ˆë‹¤.
    $$y_{1,t} = c_1 + \phi_{11,1}y_{1,t-1} + \phi_{12,1}y_{2,t-1} + \epsilon_{1,t}$$
    $$y_{2,t} = c_2 + \phi_{21,1}y_{1,t-1} + \phi_{22,1}y_{2,t-1} + \epsilon_{2,t}$$
  * **ì •ìƒì„±**: VAR ëª¨ë¸ ì—­ì‹œ ëª¨ë“  ì‹œê³„ì—´ ë°ì´í„°ê°€ ì •ìƒì„±ì„ ë§Œì¡±í•´ì•¼ í•©ë‹ˆë‹¤. ë”°ë¼ì„œ ëª¨ë¸ë§ ì „ì— ê° ë³€ìˆ˜ì— ëŒ€í•´ ADF ê²€ì • ë° ì°¨ë¶„ì„ ìˆ˜í–‰í•´ì•¼ í•©ë‹ˆë‹¤.
  * **ìµœì  ì‹œì°¨(p) ê²°ì •**: AIC, BIC ë“±ì˜ ì •ë³´ ê¸°ì¤€ì„ ì‚¬ìš©í•˜ì—¬ ìµœì ì˜ ì‹œì°¨ `p`ë¥¼ ê²°ì •í•©ë‹ˆë‹¤.

**VARMA, VARMAX ë€?**

  * **VARMA**: VAR ëª¨ë¸ì— ì´ë™í‰ê· (MA) í•­ì„ ì¶”ê°€í•˜ì—¬, ê° ë³€ìˆ˜ì˜ ì˜ˆì¸¡ ì˜¤ì°¨ ë˜í•œ ë‹¤ë¥¸ ë³€ìˆ˜ë“¤ì—ê²Œ ì˜í–¥ì„ ë¯¸ì¹˜ëŠ” ê´€ê³„ë¥¼ ëª¨ë¸ë§í•©ë‹ˆë‹¤.
  * **VARMAX**: VARMA ëª¨ë¸ì— ì™¸ìƒë³€ìˆ˜(eXogenous variable, X)ë¥¼ ì¶”ê°€í•œ ëª¨ë¸ì…ë‹ˆë‹¤. ëª¨ë¸ ì™¸ë¶€ì—ì„œ ê²°ì •ë˜ì§€ë§Œ ëª¨ë¸ ë‚´ ë³€ìˆ˜ë“¤ì—ê²Œ ì˜í–¥ì„ ì£¼ëŠ” ë³€ìˆ˜(ì˜ˆ: ì •ì±…, ì´ë²¤íŠ¸)ë¥¼ ë¶„ì„ì— í¬í•¨ì‹œí‚¬ ìˆ˜ ìˆìŠµë‹ˆë‹¤.

### ğŸ’» ì˜ˆì‹œ ì½”ë“œ (Example Code)

`moneystock-spending-series.csv` ë°ì´í„°ì˜ 'Money'ì™€ 'Spending' ë‘ ë³€ìˆ˜ê°€ ì„œë¡œì—ê²Œ ë¯¸ì¹˜ëŠ” ì˜í–¥ì„ VAR ëª¨ë¸ë¡œ ë¶„ì„í•´ ë³´ê² ìŠµë‹ˆë‹¤.

```
import pandas as pd
import plotly.graph_objects as go
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller

# ë°ì´í„° ë¡œë“œ ë° 2ì°¨ ì°¨ë¶„ (ì •ìƒì„± í™•ë³´)
df_money = pd.read_csv('https://raw.githubusercontent.com/datamanim/datarepo/main/tsa/moneystock-spending-series.csv', index_col=0, parse_dates=True)
df_money.index.freq = 'MS'
df_diff2 = df_money.diff().diff().dropna()

# í•™ìŠµ ë°ì´í„°ì™€ í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¶„ë¦¬
nobs = 12 # ë§ˆì§€ë§‰ 12ê°œë¥¼ í…ŒìŠ¤íŠ¸ ë°ì´í„°ë¡œ ì‚¬ìš©
train_var = df_diff2[:-nobs]
test_var = df_diff2[-nobs:]

# 1. ìµœì ì˜ ì‹œì°¨ p ì°¾ê¸°
model_var = VAR(train_var)
lag_selection_results = model_var.select_order(maxlags=15)
print(lag_selection_results.summary())
# AIC ê¸°ì¤€ ìµœì  ì‹œì°¨ëŠ” 5ë¡œ ë‚˜íƒ€ë‚¨

# 2. VAR(5) ëª¨ë¸ êµ¬ì¶• ë° í•™ìŠµ
model_fitted = model_var.fit(5)
print(model_fitted.summary())

# 3. ì˜ˆì¸¡
lag_order = model_fitted.k_ar
forecast_input = train_var.values[-lag_order:]
forecast_var = model_fitted.forecast(y=forecast_input, steps=nobs)
df_forecast = pd.DataFrame(forecast_var, index=test_var.index, columns=train_var.columns + '_2d')

# 4. ì˜ˆì¸¡ ê²°ê³¼ ì—­ë³€í™˜ (Invert Differencing)
def invert_transformation(df_train, df_forecast, second_diff=False):
    """ì°¨ë¶„ëœ ì˜ˆì¸¡ê°’ì„ ì›ë³¸ ìŠ¤ì¼€ì¼ë¡œ ë³µì›í•©ë‹ˆë‹¤."""
    df_fc = df_forecast.copy()
    columns = df_train.columns
    for col in columns:
        # 1ì°¨ ì°¨ë¶„ ë³µì›
        df_fc[str(col)+'_1d'] = (df_train[col].iloc[-1] - df_train[col].iloc[-2]) + df_fc[str(col)+'_2d'].cumsum()
        # 2ì°¨ ì°¨ë¶„ ë³µì›
        df_fc[str(col)+'_forecast'] = df_train[col].iloc[-1] + df_fc[str(col)+'_1d'].cumsum()
    return df_fc

df_results = invert_transformation(df_money[['Money', 'Spending']], df_forecast)

# 5. ê²°ê³¼ ì‹œê°í™”
fig = go.Figure()
# Money ë³€ìˆ˜
fig.add_trace(go.Scatter(x=df_money.index, y=df_money['Money'], name='Money Actual'))
fig.add_trace(go.Scatter(x=df_results.index, y=df_results['Money_forecast'], name='Money Forecast', line=dict(dash='dot')))
# Spending ë³€ìˆ˜
fig.add_trace(go.Scatter(x=df_money.index, y=df_money['Spending'], name='Spending Actual'))
fig.add_trace(go.Scatter(x=df_results.index, y=df_results['Spending_forecast'], name='Spending Forecast', line=dict(dash='dot')))

fig.update_layout(title='M2 í†µí™”ëŸ‰ ë° ê°œì¸ ì†Œë¹„ ì§€ì¶œ ì˜ˆì¸¡ (VAR)', xaxis_title='ë‚ ì§œ')
fig.show()
```

### âœï¸ ì—°ìŠµ ë¬¸ì œ (Practice Problems)

1.  ìºë‚˜ë‹¤ì˜ ê±°ì‹œê²½ì œ ë°ì´í„°(ê³ ìš©, ìƒì‚°ì„±, ì‹¤ì—…ë¥ , ì¸êµ¬)ë¥¼ ì‚¬ìš©í•˜ì—¬ VAR ëª¨ë¸ì„ êµ¬ì¶•í•´ë³´ì„¸ìš”.
      * ë°ì´í„°ì…‹ URL: `https://raw.githubusercontent.com/statsmodels/statsmodels/main/statsmodels/datasets/macrodata/macrodata.csv`
      * ë°ì´í„°ì…‹ì€ ë¶„ê¸°ë³„ ë°ì´í„°ì…ë‹ˆë‹¤. 'year'ì™€ 'quarter' ì—´ì„ ì¡°í•©í•˜ì—¬ ì‹œê³„ì—´ ì¸ë±ìŠ¤ë¥¼ ë§Œë“œì„¸ìš”.
