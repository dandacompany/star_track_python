```python
from langchain.embeddings import LlamaCppEmbeddings
```

```python
# 제가 양자화 처리(모델의 파라미터 수를 줄여 메모리 사용량을 줄이고 속도를 빠르게 하는 과정) 를 한 모델입니다.
# 여러분들은 직접처리하기엔 난이도가 있으니, 제가 제공해드리는 Llama2,3 모델을 사용하시면 됩니다.
llama_model_path = "./models/ggml-model-q4_0.gguf"  # 라마2
embeddings = LlamaCppEmbeddings(model_path=llama_model_path, n_ctx=2048)
```

```python
test_string = "Hi"
test_string_embedding = embeddings.embed_query(test_string)
```
