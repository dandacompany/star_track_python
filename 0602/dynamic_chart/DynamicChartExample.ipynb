```python
import warnings
warnings.filterwarnings('ignore')
import koreanize_matplotlib
```

```python
import bar_chart_race as bcr
import pandas as pd
df = pd.read_csv('../../datasets/합계출산율.csv')
df.head()
```

```python
reshape_df = df.set_index('시도').T.fillna(0)
reshape_df.head()
```

```python
reshape_df = reshape_df.replace("-", 0).fillna(0)
reshape_df = reshape_df.astype(float)
reshape_df.head()
```

```python
bcr.bar_chart_race(
        df=reshape_df, 
        orientation='h', 
        sort='desc', 
        n_bars=8, 
        fixed_order=False, 
        fixed_max=True, 
        steps_per_period=20, 
        period_length=500, 
        interpolate_period=False, 
        period_label={'x': .98, 'y': .3, 'ha': 'right', 'va': 'center'}, 
        period_summary_func=lambda v, r: {'x': .98, 'y': .2, 
                                          's': f'평균 합계출산율: {v.mean():,.3f}', 
                                          'ha': 'right', 'size': 11}, 
        perpendicular_bar_func='median', 
        title='연도별 합계출산율 비교', 
        bar_size=.95, 
        scale='linear', 
        fig=None, 
        writer=None, 
        bar_kwargs={'alpha': .7},
        filter_column_colors=False,
        )
```
