### 1. Bag-of-Words (BoW)ì™€ ë¬¸ì„œ-ë‹¨ì–´ í–‰ë ¬ (DTM)

#### ğŸ’¡ ê°œë… (Concept)

ë¨¸ì‹ ëŸ¬ë‹ ëª¨ë¸ì€ ìˆ«ì ë°ì´í„°ë§Œ ì²˜ë¦¬í•  ìˆ˜ ìˆìœ¼ë¯€ë¡œ, í…ìŠ¤íŠ¸ë¥¼ ìˆ«ì ë²¡í„°ë¡œ ë³€í™˜í•˜ëŠ” ê³¼ì •ì´ í•„ìˆ˜ì ì…ë‹ˆë‹¤. 

**Bag-of-Words(BoW)** ëŠ” ê°€ì¥ ê¸°ì´ˆì ì¸ í…ìŠ¤íŠ¸ í‘œí˜„ ë°©ë²•ìœ¼ë¡œ, ì´ë¦„ì²˜ëŸ¼ ë¬¸ì„œë¥¼ 'ë‹¨ì–´ë“¤ì˜ ê°€ë°©'ìœ¼ë¡œ ê°„ì£¼í•©ë‹ˆë‹¤. 

ì´ ëª¨ë¸ì€ ë¬¸ë§¥ì´ë‚˜ ë‹¨ì–´ì˜ ìˆœì„œëŠ” ë¬´ì‹œí•˜ê³ , ê° ë‹¨ì–´ê°€ ë¬¸ì„œì— ëª‡ ë²ˆ ë“±ì¥í–ˆëŠ”ì§€(ë¹ˆë„)ì—ë§Œ ì§‘ì¤‘í•©ë‹ˆë‹¤.

BoW ëª¨ë¸ì„ ì‚¬ìš©í•´ í…ìŠ¤íŠ¸ ëª¨ìŒ(Corpus)ì„ ë²¡í„°í™”í•˜ë©´ **ë¬¸ì„œ-ë‹¨ì–´ í–‰ë ¬(Document-Term Matrix, DTM)** ì´ ìƒì„±ë©ë‹ˆë‹¤. DTMì—ì„œ ê° í–‰ì€ ë¬¸ì„œë¥¼, ê° ì—´ì€ ì „ì²´ ì–´íœ˜ ì‚¬ì „ì— í¬í•¨ëœ ë‹¨ì–´ë¥¼, ê·¸ë¦¬ê³  ê° ì…€ì˜ ê°’ì€ í•´ë‹¹ ë‹¨ì–´ê°€ í•´ë‹¹ ë¬¸ì„œì— ë‚˜íƒ€ë‚œ ë¹ˆë„ë¥¼ ì˜ë¯¸í•©ë‹ˆë‹¤.

#### ğŸ’» ì˜ˆì‹œ ì½”ë“œ (Example Code)

`scikit-learn`ì˜ `CountVectorizer`ëŠ” BoW ëª¨ë¸ì„ ì†ì‰½ê²Œ êµ¬í˜„í•˜ë„ë¡ ë„ì™€ì¤ë‹ˆë‹¤. í•œêµ­ì–´ëŠ” ì¡°ì‚¬ê°€ ë°œë‹¬í•˜ì—¬ "ì˜í™”ê°€", "ì˜í™”ëŠ”" ë“±ì„ ë‹¤ë¥¸ ë‹¨ì–´ë¡œ ì¸ì‹í•˜ëŠ” ë¬¸ì œê°€ ìˆìœ¼ë¯€ë¡œ, `kiwipiepy` í˜•íƒœì†Œ ë¶„ì„ê¸°ë¥¼ í† í¬ë‚˜ì´ì €ë¡œ ì—°ê²°í•˜ì—¬ ì‚¬ìš©í•´ì•¼ í•©ë‹ˆë‹¤.

```python
import pandas as pd
from kiwipiepy import Kiwi
from sklearn.feature_extraction.text import CountVectorizer

# 1. kiwipiepy í˜•íƒœì†Œ ë¶„ì„ê¸° ì¤€ë¹„


# 2. CountVectorizerì— ì—°ê²°í•  í† í¬ë‚˜ì´ì € í•¨ìˆ˜ ì •ì˜
# ëª…ì‚¬(NNG, NNP), ë™ì‚¬ ì–´ê·¼(VV), í˜•ìš©ì‚¬ ì–´ê·¼(VA)ë§Œ ì¶”ì¶œ
def kiwi_tokenizer(text: str) -> list[str]:
    """kiwipiepyë¥¼ ì‚¬ìš©í•˜ì—¬ ëª…ì‚¬, ë™ì‚¬, í˜•ìš©ì‚¬ë¥¼ ì¶”ì¶œí•˜ëŠ” í† í¬ë‚˜ì´ì €"""
    kiwi = Kiwi()
    tokens = kiwi.tokenize(text)
    return [token.form for token in tokens if token.tag in ['NNG', 'NNP', 'VV', 'VA']]
```

```python
# 3. ì‹¤ìŠµìš© í…ìŠ¤íŠ¸ ë°ì´í„°
corpus = [
    'ë°°ìš°ì˜ ì—°ê¸°ë ¥ì´ ì •ë§ ëŒ€ë‹¨í•œ ì˜í™”ì˜€ì–´ìš”.',
    'ìŠ¤í† ë¦¬ê°€ ë„ˆë¬´ ì˜ˆì¸¡ ê°€ëŠ¥í•´ì„œ ì—°ê¸°ë ¥ì´ ì•„ê¹Œì› ë‹¤.',
    'ê°ë…ì˜ ì—°ì¶œê³¼ ë°°ìš°ì˜ ì—°ê¸°ê°€ ì¡°í™”ë¡œì› ë˜ ì˜í™”.',
]

# 4. CountVectorizer ìƒì„± ë° DTM êµ¬ì¶•
vectorizer = CountVectorizer(tokenizer=kiwi_tokenizer)
```

### ğŸš€ ì—¬ê¸°ì„œ ì ê¹! Python ì‹¤ë ¥ ë‹¤ì§€ê¸°: ì¢‹ì€ ë””ìì¸ íŒ¨í„´ í´ë˜ìŠ¤ë¡œ ë˜í•‘í•˜ê¸°

ìœ„ì˜ ì½”ë“œëŠ” ê°„ë‹¨í•œ ì˜ˆì‹œì˜€ì§€ë§Œ, ì‹¤ì œ í”„ë¡œì íŠ¸ì—ì„œëŠ” ë” ì²´ê³„ì ì´ê³  ì¬ì‚¬ìš© ê°€ëŠ¥í•œ ì½”ë“œ êµ¬ì¡°ê°€ í•„ìš”í•©ë‹ˆë‹¤. 

ì•„ë˜ì—ì„œëŠ” **ì‹±ê¸€í†¤ íŒ¨í„´(Singleton Pattern)** ì„ í™œìš©í•˜ì—¬ Kiwi í˜•íƒœì†Œ ë¶„ì„ê¸°ë¥¼ íš¨ìœ¨ì ìœ¼ë¡œ ê´€ë¦¬í•˜ê³ , í…ìŠ¤íŠ¸ ë§ˆì´ë‹ ì‘ì—…ì„ ìº¡ìŠí™”í•œ í´ë˜ìŠ¤ë¥¼ ë§Œë“¤ì–´ë³´ê² ìŠµë‹ˆë‹¤.

**ì£¼ìš” ê°œì„  ì‚¬í•­:**
- ğŸ¯ **ì‹±ê¸€í†¤ íŒ¨í„´**: Kiwi ê°ì²´ë¥¼ í•œ ë²ˆë§Œ ìƒì„±í•˜ì—¬ ë©”ëª¨ë¦¬ íš¨ìœ¨ì„± ì¦ëŒ€
- ğŸ”§ **ìº¡ìŠí™”**: í† í°í™”ì™€ DTM ìƒì„± ê¸°ëŠ¥ì„ í•˜ë‚˜ì˜ í´ë˜ìŠ¤ë¡œ í†µí•©
- ğŸ¨ **ìœ ì—°ì„±**: í’ˆì‚¬ íƒœê·¸ë¥¼ ë§¤ê°œë³€ìˆ˜ë¡œ ë°›ì•„ ë‹¤ì–‘í•œ ë¶„ì„ ìš”êµ¬ì‚¬í•­ ëŒ€ì‘
- ğŸ“Š **í¸ì˜ì„±**: DTMì„ ë°”ë¡œ DataFrame í˜•íƒœë¡œ ë°˜í™˜í•˜ì—¬ ë¶„ì„ ì‘ì—… ê°„ì†Œí™”

```python
import pandas as pd
from kiwipiepy import Kiwi
from sklearn.feature_extraction.text import CountVectorizer


class KiwiTokenizer:
    """Kiwi í˜•íƒœì†Œ ë¶„ì„ê¸°ë¥¼ ì‹±ê¸€í†¤ íŒ¨í„´ìœ¼ë¡œ ê´€ë¦¬í•˜ëŠ” í† í¬ë‚˜ì´ì € í´ë˜ìŠ¤"""
    
    _instance = None
    _kiwi = None
    
    def __new__(cls): # __new__ ë©”ì„œë“œëŠ” í´ë˜ìŠ¤ ì¸ìŠ¤í„´ìŠ¤ë¥¼ ìƒì„±í•  ë•Œ í˜¸ì¶œë˜ëŠ” ë©”ì„œë“œì…ë‹ˆë‹¤.
        if cls._instance is None:
            cls._instance = super(KiwiTokenizer, cls).__new__(cls)
            cls._kiwi = Kiwi() # Kiwi ê°ì²´ ìƒì„±
            cls._instance.vectorizer = None  # vectorizer ì´ˆê¸°í™”
        return cls._instance
    
    def tokenize(self, text: str, pos_tags: list[str] = None) -> list[str]:
        """
        í…ìŠ¤íŠ¸ë¥¼ í† í°í™”í•˜ì—¬ ì§€ì •ëœ í’ˆì‚¬ì˜ ë‹¨ì–´ë§Œ ì¶”ì¶œ
        
        Args:
            text: ë¶„ì„í•  í…ìŠ¤íŠ¸
            pos_tags: ì¶”ì¶œí•  í’ˆì‚¬ íƒœê·¸ ë¦¬ìŠ¤íŠ¸ (ê¸°ë³¸ê°’: ['NNG', 'NNP', 'VV', 'VA'])
        
        Returns:
            ì¶”ì¶œëœ ë‹¨ì–´ë“¤ì˜ ë¦¬ìŠ¤íŠ¸
        """
        if pos_tags is None:
            pos_tags = ['NNG', 'NNP', 'VV', 'VA']
        
        tokens = self._kiwi.tokenize(text)
        return [token.form for token in tokens if token.tag in pos_tags]
    
    def create_dtm(self, corpus: list[str], pos_tags: list[str] = None) -> pd.DataFrame:
        """
        í…ìŠ¤íŠ¸ ì½”í¼ìŠ¤ë¥¼ ì´ìš©í•˜ì—¬ DTM(Document-Term Matrix)ì„ ìƒì„±
        
        Args:
            corpus: í…ìŠ¤íŠ¸ ë¬¸ì„œë“¤ì˜ ë¦¬ìŠ¤íŠ¸
            pos_tags: ì¶”ì¶œí•  í’ˆì‚¬ íƒœê·¸ ë¦¬ìŠ¤íŠ¸ (ê¸°ë³¸ê°’: ['NNG', 'NNP', 'VV', 'VA'])
        
        Returns:
            DTM DataFrame
        """
        def tokenizer_func(text: str) -> list[str]:
            return self.tokenize(text, pos_tags)
        
        self.vectorizer = CountVectorizer(tokenizer=tokenizer_func)
        dtm = self.vectorizer.fit_transform(corpus)
        feature_names = self.vectorizer.get_feature_names_out()
        dtm_df = pd.DataFrame(dtm.toarray(), columns=feature_names)
        
        return dtm_df
    
    def get_vocabulary(self) -> list[str]:
        """ìƒì„±ëœ ì–´íœ˜ ì‚¬ì „ì„ ë°˜í™˜"""
        if self.vectorizer is None:
            print("ë¨¼ì € create_dtm() ë©”ì„œë“œë¥¼ í˜¸ì¶œí•˜ì—¬ vectorizerë¥¼ ìƒì„±í•´ì£¼ì„¸ìš”.")
            return
        feature_names = self.vectorizer.get_feature_names_out()
        return list(feature_names)
```

### 2\. TF-IDF (Term Frequency-Inverse Document Frequency)

  - **IDF (ì—­ë¬¸ì„œ ë¹ˆë„, Inverse Document Frequency)**: íŠ¹ì • ë‹¨ì–´ê°€ ì „ì²´ ë¬¸ì„œì—ì„œ ì–¼ë§ˆë‚˜ í¬ê·€í•˜ê²Œ ë“±ì¥í•˜ëŠ”ì§€ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ê°’ì…ë‹ˆë‹¤. ëª¨ë“  ë¬¸ì„œì— í”í•˜ê²Œ ë“±ì¥í•˜ëŠ” ë‹¨ì–´ëŠ” ë‚®ì€ IDF ê°’ì„, íŠ¹ì • ì†Œìˆ˜ì˜ ë¬¸ì„œì—ë§Œ ë“±ì¥í•˜ëŠ” í¬ê·€í•œ ë‹¨ì–´ëŠ” ë†’ì€ IDF ê°’ì„ ê°–ìŠµë‹ˆë‹¤.

**TF-IDF ê°’ì€ TFì™€ IDFë¥¼ ê³±í•˜ì—¬ ê³„ì‚°**ë˜ë©°, íŠ¹ì • ë¬¸ì„œì—ëŠ” ìì£¼ ë“±ì¥í•˜ì§€ë§Œ(ë†’ì€ TF) ì „ì²´ ë¬¸ì„œì—ì„œëŠ” í¬ê·€í•œ(ë†’ì€ IDF) ë‹¨ì–´ì¼ìˆ˜ë¡ ë†’ì€ ê°’ì„ ê°€ì§‘ë‹ˆë‹¤. ì´ëŠ” ê·¸ ë‹¨ì–´ê°€ í•´ë‹¹ ë¬¸ì„œë¥¼ ëŒ€í‘œí•˜ëŠ” **í•µì‹¬ì–´**ì¼ ê°€ëŠ¥ì„±ì´ ë†’ë‹¤ëŠ” ì˜ë¯¸ì…ë‹ˆë‹¤.

## 3. BM25 ì•Œê³ ë¦¬ì¦˜ (Best Matching 25)

**BM25(Best Matching 25)** ëŠ” ì •ë³´ ê²€ìƒ‰ ë¶„ì•¼ì—ì„œ ë¬¸ì„œì™€ ì§ˆì˜(query) ê°„ì˜ ê´€ë ¨ì„±ì„ í‰ê°€í•˜ëŠ” ëŒ€í‘œì ì¸ ë­í‚¹ í•¨ìˆ˜ì…ë‹ˆë‹¤.

ê²€ìƒ‰ ì—”ì§„(ì˜ˆ: êµ¬ê¸€, ë„¤ì´ë²„, ì—˜ë¼ìŠ¤í‹±ì„œì¹˜ ë“±)ì—ì„œ ì‚¬ìš©ìê°€ ì…ë ¥í•œ í‚¤ì›Œë“œì™€ ê°€ì¥ ê´€ë ¨ì„±ì´ ë†’ì€ ë¬¸ì„œë¥¼ ì°¾ì•„ ìˆœìœ„ë¥¼ ë§¤ê¸°ëŠ” ë° ë„ë¦¬ í™œìš©ë©ë‹ˆë‹¤.


### BM25ì˜ í•µì‹¬ ì›ë¦¬

- **TF(ë‹¨ì–´ ë¹ˆë„, Term Frequency):** íŠ¹ì • ë‹¨ì–´ê°€ ë¬¸ì„œ ë‚´ì—ì„œ ì–¼ë§ˆë‚˜ ìì£¼ ë“±ì¥í•˜ëŠ”ì§€ ì¸¡ì •í•©ë‹ˆë‹¤.
- **IDF(ì—­ë¬¸ì„œ ë¹ˆë„, Inverse Document Frequency):** íŠ¹ì • ë‹¨ì–´ê°€ ì „ì²´ ë¬¸ì„œ ì§‘í•©ì—ì„œ ì–¼ë§ˆë‚˜ í¬ê·€í•œì§€ ë°˜ì˜í•©ë‹ˆë‹¤. ìì£¼ ë“±ì¥í•˜ì§€ ì•ŠëŠ” ë‹¨ì–´ì¼ìˆ˜ë¡ ë” ì¤‘ìš”í•œ ì •ë³´ë¡œ ê°„ì£¼í•©ë‹ˆë‹¤.

BM25ëŠ” TF-IDFì™€ ë¹„ìŠ·í•œ ê°œë…ì„ ì‚¬ìš©í•˜ì§€ë§Œ, TF(ë‹¨ì–´ ë¹ˆë„)ì˜ ì¦ê°€ì— ë”°ë¼ ì ìˆ˜ê°€ ë¬´í•œì • ì»¤ì§€ì§€ ì•Šë„ë¡ ì¡°ì ˆí•˜ê³ , ë¬¸ì„œ ê¸¸ì´ ì°¨ì´ë„ ê³µì •í•˜ê²Œ ë°˜ì˜í•©ë‹ˆë‹¤.

### BM25ì˜ ìˆ˜ì‹

BM25 ì ìˆ˜ëŠ” ë‹¤ìŒê³¼ ê°™ì´ ê³„ì‚°ë©ë‹ˆë‹¤.

$$
\text{Score}(D, Q) = \sum_{q_i \in Q} \text{IDF}(q_i) \cdot \frac{TF(q_i, D) \cdot (k_1 + 1)}{TF(q_i, D) + k_1 \cdot (1 - b + b \cdot \frac{|D|}{\text{avgdl}})}
$$

- $TF(q_i, D)$: ë¬¸ì„œ Dì—ì„œ ë‹¨ì–´ $q_i$ì˜ ë¹ˆë„
- $|D|$: ë¬¸ì„œ Dì˜ ê¸¸ì´(ë‹¨ì–´ ìˆ˜)
- $\text{avgdl}$: ì „ì²´ ë¬¸ì„œì˜ í‰ê·  ê¸¸ì´
- $k_1, b$: ì¡°ì • ê°€ëŠ¥í•œ í•˜ì´í¼íŒŒë¼ë¯¸í„°(ë³´í†µ $k_1$ì€ 1.2~2.0, $b$ëŠ” 0.75)

```python
from kiwipiepy import Kiwi
from gensim.models import Word2Vec

# 1. ëª¨ë¸ í•™ìŠµì„ ìœ„í•œ ë°ì´í„° ì¤€ë¹„ (í† í°í™”ëœ ë¬¸ì¥ ë¦¬ìŠ¤íŠ¸)
# ì‹¤ì œ í”„ë¡œì íŠ¸ì—ì„œëŠ” ëŒ€ìš©ëŸ‰ì˜ í…ìŠ¤íŠ¸ ë°ì´í„°ê°€ í•„ìš”í•©ë‹ˆë‹¤.
# corpus = [
#     'ë°°ìš°ì˜ ì—°ê¸°ë ¥ì´ ì •ë§ ëŒ€ë‹¨í•œ ì˜í™”ì˜€ì–´ìš”',
#     'ìŠ¤í† ë¦¬ê°€ ë„ˆë¬´ ì˜ˆì¸¡ ê°€ëŠ¥í•´ì„œ ì—°ê¸°ë ¥ì´ ì•„ê¹Œì› ë‹¤',
#     'ê°ë…ì˜ ì—°ì¶œê³¼ ë°°ìš°ì˜ ì—°ê¸°ê°€ ì¡°í™”ë¡œì› ë˜ ì˜í™”',
#     'ì´ ì˜í™”ì˜ ë°°ìš°ë“¤ì€ ì—°ê¸°ë¥¼ ì •ë§ ì˜í•œë‹¤',
#     'ìŠ¤í† ë¦¬ êµ¬ì„±ì´ íƒ„íƒ„í•´ì„œ ì¢‹ì•˜ë˜ ì‘í’ˆì´ë‹¤',
#     'ì—°ì¶œì´ ì•„ì‰¬ì› ì§€ë§Œ ë°°ìš°ë“¤ì˜ ì—°ê¸°ëŠ” ìµœê³ ì˜€ë‹¤',
# ]

kiwi = Kiwi()
# ëª…ì‚¬, ë™ì‚¬, í˜•ìš©ì‚¬ë§Œ ì¶”ì¶œ
tokenized_corpus = [
    [token.form for token in kiwi.tokenize(doc) if token.tag in ['NNG', 'NNP', 'VV', 'VA']]
    for doc in corpus
]
print("í† í°í™”ëœ ë°ì´í„° (ì¼ë¶€):", tokenized_corpus[:2])
```

```python
# 2. Word2Vec ëª¨ë¸ í•™ìŠµ
# vector_size: ì„ë² ë”© ë²¡í„°ì˜ ì°¨ì›
# window: í•™ìŠµ ì‹œ ê³ ë ¤í•  ì£¼ë³€ ë‹¨ì–´ì˜ ê°œìˆ˜
# min_count: í•™ìŠµì— ì‚¬ìš©í•  ë‹¨ì–´ì˜ ìµœì†Œ ë¹ˆë„
# sg=1: Skip-gram ë°©ì‹ ì‚¬ìš© (0ì€ CBOW)
model = Word2Vec(sentences=tokenized_corpus, vector_size=100, window=3, min_count=1, sg=1)
```
