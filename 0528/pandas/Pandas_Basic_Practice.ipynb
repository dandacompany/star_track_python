  * **`.loc` (Label-based selection)**: ë ˆì´ë¸”(ì¸ë±ìŠ¤ëª…, ì»¬ëŸ¼ëª…)ì„ ê¸°ë°˜ìœ¼ë¡œ ë°ì´í„°ë¥¼ ì„ íƒí•©ë‹ˆë‹¤.

      * `df.loc[row_label_slice, col_label_slice]`
      * `df.loc[boolean_array_for_rows, boolean_array_for_cols]`
  * **`.iloc` (Integer-location based selection)**: ì •ìˆ˜ ìœ„ì¹˜(0ë¶€í„° ì‹œì‘í•˜ëŠ” ì¸ë±ìŠ¤)ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ë°ì´í„°ë¥¼ ì„ íƒí•©ë‹ˆë‹¤.

      * `df.query('ì»¬ëŸ¼ëª… > ê°’ and ì»¬ëŸ¼ëª…2 == "ë¬¸ìì—´"')`

ì‹¤ì œ ë°ì´í„°ëŠ” ì¢…ì¢… ë¹„ì–´ìˆëŠ” ê°’(ê²°ì¸¡ê°’, `NaN`)ì„ í¬í•¨í•˜ê±°ë‚˜, ë°ì´í„° íƒ€ì…ì´ ë¶„ì„ì— ì í•©í•˜ì§€ ì•Šì€ í˜•íƒœë¡œ ë˜ì–´ ìˆì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ë˜í•œ, ê¸°ì¡´ ì»¬ëŸ¼ì„ ë°”íƒ•ìœ¼ë¡œ ìƒˆë¡œìš´ ì˜ë¯¸ë¥¼ ê°€ì§„ ì»¬ëŸ¼ì„ ìƒì„±í•´ì•¼ í•  ë•Œë„ ìˆìŠµë‹ˆë‹¤.

  * **ê²°ì¸¡ê°’ ì²˜ë¦¬**:
      * í™•ì¸: `df.isnull().sum()` (ì»¬ëŸ¼ë³„ ê²°ì¸¡ê°’ ê°œìˆ˜), `df.isna().sum()`
      * ì œê±°: `df.dropna(axis=0/1, how='any'/'all', subset=['ì»¬ëŸ¼ëª…'])`
          * `axis=0`: ê²°ì¸¡ê°’ì´ ìˆëŠ” í–‰ ì œê±° (ê¸°ë³¸ê°’)
          * `axis=1`: ê²°ì¸¡ê°’ì´ ìˆëŠ” ì—´ ì œê±°
          * `how='any'`: í•˜ë‚˜ë¼ë„ ìˆìœ¼ë©´ ì œê±°, `how='all'`: ëª¨ë“  ê°’ì´ ê²°ì¸¡ê°’ì´ë©´ ì œê±°
          * `subset`: íŠ¹ì • ì»¬ëŸ¼ë“¤ì„ ê¸°ì¤€ìœ¼ë¡œ ê²°ì¸¡ê°’ ê²€ì‚¬
      * ì±„ìš°ê¸°: `df.fillna(value, method='ffill'/'bfill', limit=n)`

  * **ë°ì´í„° íƒ€ì… ë³€í™˜**: `df['ì»¬ëŸ¼ëª…'].astype(ìƒˆíƒ€ì…)` (ì˜ˆ: `int`, `float`, `str`, `category`)

      * `df.apply(í•¨ìˆ˜, axis=0/1)` (axis=0: ì—´ ë‹¨ìœ„, axis=1: í–‰ ë‹¨ìœ„)
      * `df['ì»¬ëŸ¼ëª…'].apply(í•¨ìˆ˜)`

````python
# 'salary_str'ì„ ìˆ«ìí˜•ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def convert_salary(salary):
    if pd.isna(salary) or str(salary).lower() == 'none':
        return 0
    salary_val = str(salary).lower().replace('k', '000')
    try:
        return int(salary_val)
    except ValueError:
        return 0 # ë³€í™˜ ì‹¤íŒ¨ ì‹œ 0ìœ¼ë¡œ

df_m_types = df_m.copy() # ì›ë³¸ ë°ì´í„°ë¡œ ë‹¤ì‹œ ì‹œì‘
df_m_types['salary_numeric'] = df_m_types['salary_str'].apply(convert_salary)
print("\n'salary_str'ì„ 'salary_numeric'ë¡œ ë³€í™˜ í›„")
df_m_types[['name', 'salary_str', 'salary_numeric']]
````

````python
# ì ìˆ˜ë¥¼ ê¸°ì¤€ìœ¼ë¡œ Pass/Fail ê²°ì •í•˜ëŠ” í•¨ìˆ˜ ì •ì˜
def pass_fail(score):
    if pd.isna(score):
        return 'Unknown'
    return 'Pass' if score >= 80 else 'Fail'

df_m_apply = df_m_types.copy() # íƒ€ì… ë³€í™˜ëœ ë°ì´í„° ì‚¬ìš©
df_m_apply['pass_status'] = df_m_apply['score'].apply(pass_fail)
print("apply ì˜ˆì œ (pass_status ì»¬ëŸ¼ ìƒì„±)")
df_m_apply[['name', 'score', 'pass_status']]
````
